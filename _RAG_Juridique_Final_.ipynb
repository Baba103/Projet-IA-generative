```python
from llama_index.readers.file import PDFReader
def extract_with_llama(pdf_path):
    """
    Cette fonction utilise LlamaParse (Docling) pour extraire les donnÃ©es du PDF.
    Elle prend en charge les documents complexes avec des images, des logos et des tableaux.
    """
    # Charger le document PDF
    loader = PDFReader()
    document = loader.load_data(pdf_path)

    # Extraire le contenu du PDF
    content = ""
    for doc in document:
        content += doc.get_content()  # RÃ©cupÃ©rer le texte extrait
    return content

# Extraction du texte avec LlamaParse
print("Extraction avec LlamaParse en cours...")
text_llama = extract_with_llama(pdf_path)

# Sauvegarder le texte extrait dans un fichier texte
with open("/content/extracted_llama.txt", "w", encoding="utf-8") as file_llama:
    file_llama.write(text_llama)

print("Texte extrait par LlamaParse sauvegardÃ© dans 'extracted_llama.txt'.")
```

### ğŸ§© 2. Chunking Intelligent (DÃ©coupage)

**La clÃ© d'un RAG efficace** : Le texte est trop long pour Ãªtre analysÃ© en un bloc.
ğŸ‘‰ Nous le dÃ©coupons en morceaux de taille fixe (*chunks*) avec `RecursiveCharacterTextSplitter`, qui respecte la structure grammaticale pour ne pas couper les phrases au milieu.

```python
import re

def split_recursive(text):
    """
    Cette fonction dÃ©coupe le texte en morceaux ("chunks") tout en respectant la structure logique
    du document (paragraphes, phrases, etc.), en utilisant un dÃ©coupage rÃ©cursif (hiÃ©rarchique).
    """
    # 1. DÃ©couper par paragraphes
    paragraphs = text.split("\n\n")  # Supposons que les paragraphes sont sÃ©parÃ©s par des retours Ã  la ligne doubles

    chunks = []
    for paragraph in paragraphs:
        # 2. DÃ©couper les paragraphes en phrases
        sentences = paragraph.split(". ")

        # Si une phrase est trop longue, on peut la couper davantage (par exemple par mots ou caractÃ¨res)
        for sentence in sentences:
            if len(sentence) > 1000:  # Limite arbitraire pour la longueur d'un chunk
                words = sentence.split()
                for i in range(0, len(words), 50):  # DÃ©couper en morceaux de 50 mots maximum
                    chunks.append(" ".join(words[i:i+50]))
            else:
                chunks.append(sentence)  # Ajouter la phrase intacte si elle est dÃ©jÃ  assez courte

    return chunks

# Chemin du fichier texte extrait
pdf_text_path = "/content/extracted_llama.txt"

# Lire le texte du fichier
with open(pdf_text_path, "r", encoding="utf-8") as file:
    text = file.read()

# Appliquer le dÃ©coupage rÃ©cursif au texte extrait
chunks = split_recursive(text)

# Afficher les "chunks" rÃ©sultants
for i, chunk in enumerate(chunks):
    print(f"Chunk {i+1}: {chunk}\n")

# Sauvegarder les chunks dans un fichier
output_path = "/content/split_chunks.txt"
with open(output_path, "w", encoding="utf-8") as file:
    for chunk in chunks:
        file.write(chunk + "\n\n")

print(f"DÃ©coupage terminÃ© et sauvegardÃ© dans '{output_path}'.")
```

### ğŸ’¾ 3. Vectorisation & Stockage

**Transformation mathÃ©matique** : Chaque chunk est converti en vecteur.
Ces vecteurs sont stockÃ©s dans **ChromaDB**, une base de donnÃ©es optimisÃ©e pour retrouver instantanÃ©ment les textes similaires Ã  une question.

```python
import chromadb
from sentence_transformers import SentenceTransformer
import numpy as np

# Charger le modÃ¨le Sentence-Transformer
model = SentenceTransformer('all-MiniLM-L6-v2')

def embed_text(text_chunks):
    """
    Cette fonction prend une liste de chunks de texte et les transforme en vecteurs numÃ©riques.
    Elle utilise le modÃ¨le Sentence-Transformers pour gÃ©nÃ©rer des embeddings de chaque chunk.
    """
    embeddings = model.encode(text_chunks, convert_to_tensor=True)
    return embeddings

# Lire le fichier de texte extrait (assurez-vous que le fichier est correctement formatÃ©)
file_path = "/content/split_chunks.txt"
with open(file_path, "r", encoding="utf-8") as file:
    # Read all lines and strip potential newlines for cleaner chunks
    text_chunks = [line.strip() for line in file.readlines() if line.strip()]

# Appliquer l'Embedding sur les chunks de texte
embeddings = embed_text(text_chunks)

# Affichage de l'embedding du premier chunk pour vÃ©rifier
print("Embedding du premier chunk :", embeddings[0])

# Sauvegarder les embeddings dans un fichier (en format numpy pour un usage futur)
embedding_output_path = "/content/embeddings.npy"
np.save(embedding_output_path, embeddings.cpu().numpy())

print(f"Embeddings sauvegardÃ©s dans '{embedding_output_path}'.")

# --- FIX: Create and populate ChromaDB collection --- #
client = chromadb.Client()
# Create or get the collection named 'juridical_documents'
collection = client.get_or_create_collection(name="juridical_documents")

# Add the chunks and their embeddings to the ChromaDB collection
# Ensure embeddings are converted to a list for ChromaDB
if text_chunks and embeddings.shape[0] == len(text_chunks):
    collection.add(
        documents=text_chunks,
        embeddings=embeddings.cpu().numpy().tolist(), # Convert to list of lists for ChromaDB
        ids=[f"chunk_{i}" for i in range(len(text_chunks))]
    )
    print(f"ğŸ’¾ {len(text_chunks)} chunks et embeddings ajoutÃ©s Ã  la collection ChromaDB 'juridical_documents'.")
else:
    print("âš ï¸ Aucun chunk Ã  ajouter Ã  ChromaDB ou mismatch entre le nombre de chunks et d'embeddings.")
```

# ğŸ’¾ 4. RETRIEVE - La Recherche de l'Information

## MÃ©thode choisie : **Recherche par SimilaritÃ© Cosinus dans le Vector Store**
Cette Ã©tape permet de **transformer la question** de l'utilisateur en un vecteur et de la comparer avec les **vecteurs des documents** dans la base de donnÃ©es pour rÃ©cupÃ©rer les **documents les plus similaires**, afin de fournir une rÃ©ponse pertinente Ã  la question posÃ©e.



```python


# Initialiser le modÃ¨le Sentence-Transformer pour l'encoding
model = SentenceTransformer('all-MiniLM-L6-v2')

# Connexion Ã  ChromaDB
client = chromadb.Client()

# CrÃ©er ou rÃ©cupÃ©rer la collection pour les documents
collection = client.get_collection("juridical_documents")

def vectorize_question(question):
    """
    Cette fonction transforme la question en vecteur en utilisant le mÃªme modÃ¨le Bi-Encoder.
    Elle convertit ensuite le tensor en numpy array pour Ãªtre compatible avec ChromaDB.
    """
    question_vector = model.encode([question], convert_to_tensor=True)
    # Fix: Move the tensor to CPU before converting to numpy
    return question_vector.cpu().numpy()  # Convertir en numpy array aprÃ¨s l'avoir dÃ©placÃ© vers le CPU

def retrieve_relevant_chunks(question, top_k=3):
    """
    Cette fonction recherche les documents les plus similaires Ã  la question dans ChromaDB
    et retourne les top_k chunks les plus pertinents.
    """
    # Vectoriser la question
    question_vector = vectorize_question(question)

    # Recherche par similaritÃ© dans ChromaDB
    results = collection.query(
        query_embeddings=question_vector.tolist(), # ChromaDB expects a list of embeddings
        n_results=top_k
    )

    return results['documents'], results['distances']  # Les documents et leurs scores de similaritÃ©

# Exemple de question juridique
question = "Quel sont les Services du SecrÃ©tariat GÃ©nÃ©ral ?"

# Appliquer la recherche dans ChromaDB
relevant_chunks, similarities = retrieve_relevant_chunks(question)

# Afficher les rÃ©sultats
for i, (chunk_list, similarity_list) in enumerate(zip(relevant_chunks, similarities)):
    # On assume que similarity_list contient les valeurs de similaritÃ© pour chaque document
    # Formater chaque score de similaritÃ© et afficher les rÃ©sultats
    for j, (chunk, similarity) in enumerate(zip(chunk_list, similarity_list)):
        print(f"Chunk {i+1} - RÃ©sultat {j+1} (similaritÃ© = {similarity:.4f}) :\n{chunk}\n")
```

```python
from chonkie import Pipeline
from chonkie import Pipeline
# Pipeline de chunking avancÃ© (SÃ©mantique + RÃ©cursif)
pipe = (
    Pipeline()
    .chunk_with("recursive", tokenizer="roberta-base", chunk_size=2048, recipe="markdown")
    .chunk_with("semantic", chunk_size=512)
    .refine_with("overlap", context_size=128)
    .refine_with("embeddings", embedding_model="sentence-transformers/all-MiniLM-L6-v2")
)

# Extraire le texte de chaque Ã©lÃ©ment et joindre en une seule chaÃ®ne
full_text = "\n\n".join([element.text for element in document if hasattr(element, 'text')])

# ExÃ©cuter le pipeline sur le texte extrait
doc = pipe.run(texts=full_text)

print("âœ… Chunking SÃ©mantique terminÃ©.")
print(f"Exemple de chunk intelligent : \n{doc.chunks[0].text[:300]}...")
```

```python
# Code complet de la Version 4.0 (Groq Integration)

import os
import re
import json
import hashlib
import torch
from datetime import datetime
from typing import List, Dict, Tuple, Optional
from tqdm import tqdm
from unstructured.partition import pdf
from chonkie import Pipeline
from sentence_transformers import SentenceTransformer
import chromadb
from groq import Groq

# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# âš™ï¸ CONFIGURATION
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

GROQ_API_KEY = "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"
PDF_PATH = "/content/1389 bis FR.pdf"  # â† Adaptez si nÃ©cessaire

CONFIG = {
    "pdf_path": PDF_PATH,
    "collection_name": "juridical_docs_groq",
    "embedding_model": "sentence-transformers/all-MiniLM-L6-v2",
    "groq_model": "llama-3.3-70b-versatile",
    "groq_api_key": GROQ_API_KEY,
    "chunk_size": 512,
    "chunk_overlap": 128,
    "top_k_retrieval": 5,
    "device": "cuda" if torch.cuda.is_available() else "cpu"
}

# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# ğŸ“ LOGGER
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

class Logger:
    def __init__(self, log_file: str = None):
        self.log_file = log_file or "rag_log.txt"

    def log(self, message: str, level: str = "INFO"):
        timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        log_message = f"[{timestamp}] [{level}] {message}"
        print(log_message)

logger = Logger()

# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# ğŸ“„ EXTRACTION PDF
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

class DocumentExtractor:
    def __init__(self, pdf_path: str):
        self.pdf_path = pdf_path

    def extract(self) -> str:
        logger.log(f"ğŸ“„ Extraction du PDF: {self.pdf_path}", "INFO")
        document = pdf.partition_pdf(filename=self.pdf_path)
        text = " ".join([elem.text for elem in document if hasattr(elem, 'text')])
        logger.log(f"âœ… Extraction rÃ©ussie: {len(text)} caractÃ¨res", "SUCCESS")
        return text

# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# âœ‚ï¸ CHUNKING
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

class IntelligentChunker:
    def __init__(self, chunk_size: int = 512, overlap: int = 128):
        self.chunk_size = chunk_size
        self.overlap = overlap
        self.pipeline = None

    def create_pipeline(self):
        self.pipeline = (
            Pipeline()
            .chunk_with("recursive", tokenizer="roberta-base", chunk_size=2048, recipe="markdown")
            .chunk_with("semantic", chunk_size=self.chunk_size)
            .refine_with("overlap", context_size=self.overlap)
            .refine_with("embeddings", embedding_model=CONFIG["embedding_model"])
        )

    def chunk_text(self, text: str) -> List[str]:
        logger.log("âœ‚ï¸ DÃ©coupage en chunks...", "INFO")
        if not self.pipeline:
            self.create_pipeline()
        doc = self.pipeline.run(texts=text)
        chunks = [chunk.text for chunk in doc.chunks]
        logger.log(f"âœ… {len(chunks)} chunks crÃ©Ã©s", "SUCCESS")
        return chunks

# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# ğŸ“Š MÃ‰TADONNÃ‰ES
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

class MetadataExtractor:
    @staticmethod
    def extract(chunk_text: str, chunk_index: int, source_doc: str) -> Dict:
        patterns = {
            "article": r'Article\s+(\d+)',
            "decret": r'D[Ã©e]cret\s+n[Â°\s]*(\d+-\d+)',
            "loi": r'Loi\s+n[Â°\s]*(\d+-\d+)',
        }

        metadata = {
            "chunk_index": chunk_index,
            "chunk_hash": hashlib.md5(chunk_text.encode()).hexdigest(),
            "chunk_length": len(chunk_text),
            "source_document": source_doc,
        }

        for entity_type, pattern in patterns.items():
            matches = re.findall(pattern, chunk_text, re.IGNORECASE)
            if matches:
                metadata[f"has_{entity_type}"] = True
                metadata[f"{entity_type}_numbers"] = json.dumps(matches)
                metadata[f"{entity_type}_count"] = len(matches)
                metadata[f"{entity_type}_first"] = matches[0]
            else:
                metadata[f"has_{entity_type}"] = False
                metadata[f"{entity_type}_numbers"] = ""
                metadata[f"{entity_type}_count"] = 0
                metadata[f"{entity_type}_first"] = ""

        legal_terms = ["article", "dÃ©cret", "loi"]
        legal_count = sum(chunk_text.lower().count(term) for term in legal_terms)
        metadata["legal_density"] = legal_count / max(len(chunk_text.split()), 1)

        return metadata

# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# ğŸ—„ï¸ VECTOR STORE
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

class VectorStoreManager:
    def __init__(self, collection_name: str, embedding_model_name: str):
        self.collection_name = collection_name
        self.embedding_model_name = embedding_model_name
        self.client = None
        self.collection = None
        self.embedding_model = None

    def initialize(self):
        logger.log("ğŸ”Œ Initialisation de ChromaDB", "INFO")
        self.client = chromadb.Client()
        try:
            self.client.delete_collection(self.collection_name)
        except:
            pass
        self.collection = self.client.create_collection(name=self.collection_name)
        logger.log("ğŸ“¥ Chargement modÃ¨le d'embedding...", "INFO")
        self.embedding_model = SentenceTransformer(self.embedding_model_name)
        logger.log("âœ… Vector store initialisÃ©", "SUCCESS")

    def add_documents(self, chunks: List[str], source_document: str):
        logger.log(f"ğŸ“Š Ajout de {len(chunks)} documents...", "INFO")
        embeddings = self.embedding_model.encode(chunks, show_progress_bar=True, convert_to_numpy=True)

        for i, (chunk, embedding) in enumerate(tqdm(zip(chunks, embeddings), desc="Stockage", total=len(chunks))):
            metadata = MetadataExtractor.extract(chunk, i, source_document)
            self.collection.add(
                documents=[chunk],
                embeddings=[embedding.tolist()],
                metadatas=[metadata],
                ids=[f"chunk_{i}"]
            )
        logger.log("âœ… Documents ajoutÃ©s avec succÃ¨s", "SUCCESS")

# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# ğŸ” RETRIEVER
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

class HybridRetriever:
    def __init__(self, collection, embedding_model):
        self.collection = collection
        self.embedding_model = embedding_model

    def retrieve(self, question: str, top_k: int = 5) -> Tuple[List[str], List[Dict]]:
        question_embedding = self.embedding_model.encode([question], convert_to_numpy=True)
        results = self.collection.query(
            query_embeddings=question_embedding.tolist(),
            n_results=top_k * 2,
            include=['documents', 'metadatas', 'distances']
        )

        if not results['documents'][0]:
            return [], []

        scored_results = []
        for doc, metadata, distance in zip(results['documents'][0], results['metadatas'][0], results['distances'][0]):
            score = 1 / (1 + distance)
            if metadata.get('has_article'):
                score *= 1.2
            scored_results.append({
                'document': doc,
                'metadata': metadata,
                'score': score
            })

        scored_results.sort(key=lambda x: x['score'], reverse=True)
        final_results = scored_results[:top_k]
        documents = [r['document'] for r in final_results]

        logger.log(f"âœ… {len(documents)} rÃ©sultats trouvÃ©s", "SUCCESS")
        return documents, final_results

# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# ğŸ¤– GÃ‰NÃ‰RATEUR GROQ
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

class GroqGenerator:
    def __init__(self, api_key: str, model: str):
        self.client = Groq(api_key=api_key)
        self.model = model
        logger.log(f"âœ… Groq initialisÃ©: {model}", "SUCCESS")

    def generate(self, question: str, context_chunks: List[str]) -> str:
        context_str = "\n\n".join([
            f"[Source {i+1}] {chunk[:500]}" if len(chunk) > 500 else f"[Source {i+1}] {chunk}"
            for i, chunk in enumerate(context_chunks[:3])
        ])

        messages = [
            {
                "role": "system",
                "content": "Tu es un assistant juridique expert. RÃ©ponds prÃ©cisÃ©ment en te basant sur les sources. Pour les listes, utilise des points (â€¢)."
            },
            {
                "role": "user",
                "content": f"""Sources du Journal Officiel de Mauritanie:

{context_str}

QUESTION: {question}

RÃ©ponds de maniÃ¨re structurÃ©e et prÃ©cise."""
            }
        ]

        try:
            logger.log("ğŸ¤– GÃ©nÃ©ration via Groq...", "INFO")
            response = self.client.chat.completions.create(
                model=self.model,
                messages=messages,
                temperature=0.3,
                max_tokens=500
            )
            answer = response.choices[0].message.content
            logger.log("âœ… RÃ©ponse gÃ©nÃ©rÃ©e", "SUCCESS")
            return answer
        except Exception as e:
            error = f"Erreur Groq: {str(e)}"
            logger.log(error, "ERROR")
            if "api_key" in str(e).lower():
                return "âŒ Erreur: ClÃ© API invalide. VÃ©rifiez votre clÃ© sur https://console.groq.com/keys"
            return f"âŒ Erreur: {error}"

# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# ğŸ¯ SYSTÃˆME RAG COMPLET
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

class RAGSystemWithGroq:
    def __init__(self, config: Dict):
        self.config = config
        self.extractor = None
        self.chunker = None
        self.vector_store = None
        self.retriever = None
        self.generator = None

    def setup(self):
        logger.log("="*80, "INFO")
        logger.log("ğŸš€ INITIALISATION DU SYSTÃˆME RAG AVEC GROQ", "INFO")
        logger.log("="*80, "INFO")

        # 1. Extraction
        logger.log("\nğŸ“„ Ã‰TAPE 1: EXTRACTION", "INFO")
        self.extractor = DocumentExtractor(self.config["pdf_path"])
        text = self.extractor.extract()

        # 2. Chunking
        logger.log("\nâœ‚ï¸ Ã‰TAPE 2: CHUNKING", "INFO")
        self.chunker = IntelligentChunker(self.config["chunk_size"], self.config["chunk_overlap"])
        chunks = self.chunker.chunk_text(text)

        # 3. Vector Store
        logger.log("\nğŸ“Š Ã‰TAPE 3: BASE VECTORIELLE", "INFO")
        self.vector_store = VectorStoreManager(self.config["collection_name"], self.config["embedding_model"])
        self.vector_store.initialize()
        self.vector_store.add_documents(chunks, os.path.basename(self.config["pdf_path"]))

        # 4. Retriever
        logger.log("\nğŸ” Ã‰TAPE 4: RETRIEVER", "INFO")
        self.retriever = HybridRetriever(self.vector_store.collection, self.vector_store.embedding_model)

        # 5. Groq Generator
        logger.log("\nğŸŒ Ã‰TAPE 5: GROQ API", "INFO")
        self.generator = GroqGenerator(self.config["groq_api_key"], self.config["groq_model"])

        logger.log("\n"+"="*80, "INFO")
        logger.log("âœ… SYSTÃˆME RAG AVEC GROQ PRÃŠT!", "SUCCESS")
        logger.log("="*80+"\n", "INFO")

    def query(self, question: str) -> Dict:
        documents, results = self.retriever.retrieve(question, self.config["top_k_retrieval"])

        if not documents:
            return {"answer": "Aucun document pertinent trouvÃ©.", "num_sources": 0}

        answer = self.generator.generate(question, documents)

        return {
            "answer": answer,
            "num_sources": len(documents),
            "sources": [{"text": doc, "metadata": r["metadata"], "score": r["score"]} for doc, r in zip(documents, results)]
        }

# ExÃ©cution de la dÃ©mo
if os.path.exists(CONFIG["pdf_path"]):
    rag = RAGSystemWithGroq(CONFIG)
    rag.setup()

    print("\nğŸ§ª TEST V4.0 :")
    response = rag.query("Quel est le rÃ´le du Directeur ?")
    print(f"\nRÃ‰PONSE: {response['answer']}")
else:
    print("âš ï¸ Fichier PDF non trouvÃ© pour la dÃ©mo V4.0. Assurez-vous d'avoir tÃ©lÃ©chargÃ© un journal.")
```

### ğŸ§  Version 4.1 : MÃ©moire & Contextualisation

**ProblÃ¨me** : Le chatbot oubliait de quoi on parlait (ex: Q1 "Qui est le ministre ?", Q2 "Et son adjoint ?" -> Erreur).
**Solution** : ImplÃ©mentation d'une classe `ConversationMemory` (File) pour l'historique et `EntityTracker` (Regex) pour suivre les sujets (Article X, Direction Y).

Cette brique est essentielle pour passer d'un simple moteur de recherche Ã  un vÃ©ritable **Agent Conversationnel**.

```python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# ğŸ§  RAG AVEC MÃ‰MOIRE - Version 4.1
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•


CONFIG = {
    "pdf_path": PDF_PATH,
    "output_dir": "/content/rag_memoire_stable",
    "collection_name": "juridical_docs_stable",
    "embedding_model": "sentence-transformers/all-MiniLM-L6-v2",
    "groq_model": "llama-3.3-70b-versatile",
    "groq_api_key": GROQ_API_KEY,
    "chunk_size": 512,
    "chunk_overlap": 128,
    "memory_turns": 10,
    "confidence_threshold": 0.5,
}

os.makedirs(CONFIG["output_dir"], exist_ok=True)

# Logger
class Logger:
    def __init__(self):
        self.log_file = os.path.join(CONFIG["output_dir"], "log.txt")
    def log(self, msg, level="INFO"):
        ts = datetime.now().strftime("%H:%M:%S")
        log = f"[{ts}] [{level}] {msg}"
        print(log)
        with open(self.log_file, "a", encoding="utf-8") as f:
            f.write(log + "\n")

logger = Logger()

# MÃ‰MOIRE CONVERSATIONNELLE
class ConversationMemory:
    def __init__(self, max_turns=10):
        self.max_turns = max_turns
        self.history = deque(maxlen=max_turns)
        self.full_history = []

    def add(self, question, answer, metadata=None):
        turn = {
            "timestamp": datetime.now().isoformat(),
            "question": question,
            "answer": answer,
            "metadata": metadata or {}
        }
        self.history.append(turn)
        self.full_history.append(turn)

    def get_context(self, n=2):
        recent = list(self.history)[-n:]
        return "\n".join([f"Q: {t['question']}\nR: {t['answer'][:150]}" for t in recent])

    def clear(self):
        self.history.clear()

    def save(self, filepath):
        with open(filepath, 'w', encoding='utf-8') as f:
            json.dump(self.full_history, f, ensure_ascii=False, indent=2)

# ENTITY TRACKER
class EntityTracker:
    def __init__(self):
        self.entities = {"articles": set(), "decrets": set(), "directions": set()}
        self.last = {}

    def extract(self, text):
        articles = re.findall(r'[Aa]rticle\s+(\d+)', text)
        if articles:
            self.entities["articles"].update(articles)
            self.last["article"] = articles[-1]

    def resolve(self, pronoun):
        if pronoun.lower() in ["il", "elle", "le", "la"] and "article" in self.last:
            return f"Article {self.last['article']}"
        return pronoun

# COREFERENCE RESOLVER
class CoreferenceResolver:
    PRONOUNS = ["il", "elle", "le", "la", "premier", "premiÃ¨re", "dernier"]

    def __init__(self, tracker):
        self.tracker = tracker

    def resolve(self, question, memory):
        words = question.lower().split()
        for word in words:
            cleaned = word.strip('?,!.;:')
            if cleaned in self.PRONOUNS:
                resolution = self.tracker.resolve(cleaned)
                if resolution != cleaned:
                    logger.log(f"ğŸ”„ '{cleaned}' â†’ '{resolution}'")
                    question = question.replace(cleaned, resolution)
        return question

# EXTRACTION
class DocumentExtractor:
    def __init__(self, pdf_path):
        self.pdf_path = pdf_path

    def extract(self):
        logger.log(f"ğŸ“„ Extraction: {self.pdf_path}")
        doc = pdf.partition_pdf(filename=self.pdf_path)
        text = " ".join([e.text for e in doc if hasattr(e, 'text')])
        logger.log(f"âœ… {len(text)} caractÃ¨res")
        return text

# CHUNKING
class IntelligentChunker:
    def __init__(self, size=512, overlap=128):
        self.size = size
        self.overlap = overlap
        self.pipeline = None

    def create_pipeline(self):
        self.pipeline = (
            Pipeline()
            .chunk_with("recursive", tokenizer="roberta-base", chunk_size=2048, recipe="markdown")
            .chunk_with("semantic", chunk_size=self.size)
            .refine_with("overlap", context_size=self.overlap)
            .refine_with("embeddings", embedding_model=CONFIG["embedding_model"])
        )

    def chunk(self, text):
        logger.log("âœ‚ï¸ Chunking...")
        if not self.pipeline:
            self.create_pipeline()
        doc = self.pipeline.run(texts=text)
        chunks = [c.text for c in doc.chunks]
        logger.log(f"âœ… {len(chunks)} chunks")
        return chunks

# METADATA
class MetadataExtractor:
    @staticmethod
    def extract(chunk, idx, source):
        patterns = {
            "article": r'Article\s+(\d+)',
            "decret": r'D[Ã©e]cret\s+n?[Â°\s]*(\d+-\d+)',
        }
        metadata = {
            "chunk_index": idx,
            "chunk_hash": hashlib.md5(chunk.encode()).hexdigest(),
            "source_document": source,
        }
        for etype, pattern in patterns.items():
            matches = re.findall(pattern, chunk, re.IGNORECASE)
            metadata[f"has_{etype}"] = bool(matches)
            metadata[f"{etype}_numbers"] = json.dumps(matches) if matches else ""
        return metadata

# VECTOR STORE
class VectorStore:
    def __init__(self, name, embedding_model_name):
        self.name = name
        self.client = None
        self.collection = None
        self.embedding_model = None
        self.embedding_model_name = embedding_model_name

    def init(self):
        logger.log("ğŸ”Œ ChromaDB init")
        self.client = chromadb.Client()
        try:
            self.client.delete_collection(self.name)
        except:
            pass
        self.collection = self.client.create_collection(name=self.name)
        self.embedding_model = SentenceTransformer(self.embedding_model_name)
        logger.log("âœ… Vector store OK")

    def add(self, chunks, source):
        logger.log(f"ğŸ“Š Ajout {len(chunks)} docs")
        embeddings = self.embedding_model.encode(chunks, show_progress_bar=True, convert_to_numpy=True)
        for i, (chunk, emb) in enumerate(tqdm(zip(chunks, embeddings), desc="Stockage", total=len(chunks))):
            metadata = MetadataExtractor.extract(chunk, i, source)
            self.collection.add(
                documents=[chunk],
                embeddings=[emb.tolist()],
                metadatas=[metadata],
                ids=[f"chunk_{i}"]
            )
        logger.log("âœ… Docs ajoutÃ©s")

# RECHERCHE INTERNE
class InternalSearch:
    def __init__(self, collection, embedding_model):
        self.collection = collection
        self.embedding_model = embedding_model

    def search(self, query, n=10):
        emb = self.embedding_model.encode([query], convert_to_numpy=True)
        results = self.collection.query(
            query_embeddings=emb.tolist(),
            n_results=n,
            include=['documents', 'metadatas', 'distances']
        )
        if not results['documents'][0]:
            return {'results': [], 'confidence': 0.0}

        distances = results['distances'][0]
        scores = [1/(1+d) for d in distances]
        confidence = sum(scores)/len(scores) if scores else 0.0

        formatted = []
        for doc, meta, score in zip(results['documents'][0], results['metadatas'][0], scores):
            formatted.append({'text': doc, 'metadata': meta, 'score': score})

        return {'results': formatted, 'confidence': confidence}

# RECHERCHE WEB
class WebSearch:
    def __init__(self, groq_api_key):
        self.groq = Groq(api_key=groq_api_key)

    def search(self, query):
        logger.log(f"ğŸŒ Web search: {query}")
        try:
            # Simple DuckDuckGo via requests (pas de LangChain)
            url = f"https://html.duckduckgo.com/html/?q={requests.utils.quote(query)}"
            headers = {'User-Agent': 'Mozilla/5.0'}
            response = requests.get(url, headers=headers, timeout=5)
            soup = BeautifulSoup(response.text, 'html.parser')
            snippets = [r.get_text()[:200] for r in soup.find_all('a', class_='result__snippet')][:3]
            results = "\n".join(snippets) if snippets else "Aucun rÃ©sultat"

            # SynthÃ¨se avec Groq
            summary = self._synthesize(query, results)
            return {'summary': summary, 'source': 'web', 'warning': 'âš ï¸ Source externe'}
        except Exception as e:
            logger.log(f"âŒ Web error: {e}", "ERROR")
            return {'summary': f"Erreur: {e}", 'source': 'web', 'warning': 'âŒ Ã‰chec'}

    def _synthesize(self, query, results):
        try:
            response = self.groq.chat.completions.create(
                model="llama-3.3-70b-versatile",
                messages=[{"role": "user", "content": f"SynthÃ¨se:\n{results[:1500]}"}],
                max_tokens=300
            )
            return response.choices[0].message.content
        except:
            return results[:500]

# SYSTÃˆME RAG AVEC MÃ‰MOIRE
class RAGWithMemory:
    def __init__(self, config, vector_store, internal_search, web_search):
        self.config = config
        self.vector_store = vector_store
        self.internal = internal_search
        self.web = web_search
        self.groq = Groq(api_key=config["groq_api_key"])

        # MÃ©moire
        self.memory = ConversationMemory(max_turns=config["memory_turns"])
        self.entities = EntityTracker()
        self.resolver = CoreferenceResolver(self.entities)

        logger.log("âœ… RAG avec mÃ©moire OK")

    def query(self, question, use_fallback=True):
        logger.log(f"\n{'='*80}")
        logger.log(f"â“ {question}")

        # RÃ©solution corÃ©fÃ©rences
        enriched = self.resolver.resolve(question, self.memory)

        # Contexte conversationnel
        conv_ctx = self.memory.get_context(n=2)

        # Recherche interne
        logger.log("ğŸ” Recherche interne")
        internal_results = self.internal.search(enriched)
        confidence = internal_results['confidence']
        logger.log(f"ğŸ“Š Confiance: {confidence:.2f}")

        # Fallback web ?
        use_web = False
        web_results = None
        if use_fallback and confidence < self.config["confidence_threshold"]:
            logger.log("âš ï¸ Activation web")
            use_web = True
            web_results = self.web.search(enriched)

        # PrÃ©parer contexte
        contexts = []
        if internal_results['results']:
            contexts = [r['text'] for r in internal_results['results'][:5]]
        if use_web and web_results:
            contexts.append(web_results['summary'])

        if not contexts:
            answer = "Aucune information trouvÃ©e."
            self.memory.add(question, answer)
            return {'answer': answer, 'confidence': 0.0, 'used_web': False}

        # GÃ©nÃ©rer rÃ©ponse
        answer = self._generate(question, enriched, contexts, conv_ctx, use_web)

        # Mise Ã  jour mÃ©moire
        self.memory.add(question, answer, {'confidence': confidence, 'used_web': use_web})
        self.entities.extract(question + " " + answer)

        logger.log("âœ… RÃ©ponse gÃ©nÃ©rÃ©e")

        return {
            'answer': answer,
            'confidence': confidence,
            'used_web': use_web,
            'enriched': enriched != question
        }

    def _generate(self, original, enriched, contexts, conv_ctx, used_web):
        ctx_str = "\n\n".join([f"[Source {i+1}]\n{c[:500]}" for i, c in enumerate(contexts)])
        mem_note = f"\n\nCONTEXTE:\n{conv_ctx}" if conv_ctx else ""
        web_note = "\nâš ï¸ Sources externes" if used_web else ""

        prompt = f"""Assistant juridique avec mÃ©moire.

SOURCES:
{ctx_str}
{mem_note}
{web_note}

QUESTION: {original}

RÃ©ponds de maniÃ¨re structurÃ©e avec â€¢."""

        try:
            response = self.groq.chat.completions.create(
                model=self.config["groq_model"],
                messages=[{"role": "user", "content": prompt}],
                temperature=0.3,
                max_tokens=500
            )
            return response.choices[0].message.content
        except Exception as e:
            return f"Erreur: {e}"

    def reset(self):
        self.memory.clear()
        self.entities = EntityTracker()
        logger.log("ğŸ”„ MÃ©moire reset")

    def save(self):
        path = os.path.join(self.config["output_dir"], f"conv_{datetime.now().strftime('%Y%m%d_%H%M%S')}.json")
        self.memory.save(path)
        logger.log(f"ğŸ’¾ SauvegardÃ©: {path}")
        return path

# INITIALISATION
def init_system():
    logger.log("="*80)
    logger.log("ğŸ§  INIT RAG AVEC MÃ‰MOIRE")
    logger.log("="*80)

    extractor = DocumentExtractor(CONFIG["pdf_path"])
    text = extractor.extract()

    chunker = IntelligentChunker(CONFIG["chunk_size"], CONFIG["chunk_overlap"])
    chunks = chunker.chunk(text)

    store = VectorStore(CONFIG["collection_name"], CONFIG["embedding_model"])
    store.init()
    store.add(chunks, os.path.basename(CONFIG["pdf_path"]))

    internal = InternalSearch(store.collection, store.embedding_model)
    web = WebSearch(CONFIG["groq_api_key"])

    system = RAGWithMemory(CONFIG, store, internal, web)

    logger.log("\n"+"="*80)
    logger.log("âœ… SYSTÃˆME PRÃŠT!")
    logger.log("="*80+"\n")

    return system

# FONCTIONS HELPER
def poser_question(q):
    print("\n" + "="*80)
    print(f"â“ {q}")
    print("="*80)

    r = rag.query(q)

    print(f"\nğŸ’¬ RÃ‰PONSE:\n{r['answer']}")
    print(f"\nğŸ“Š Confiance: {r['confidence']:.2%}")

    if r.get('enriched'):
        print("ğŸ”„ Question enrichie")
    if r.get('used_web'):
        print("âš ï¸ Inclut sources web")

    print("="*80)
    return r

def nouvelle_conversation():
    rag.reset()
    print("ğŸ”„ Nouvelle conversation - MÃ©moire reset")

def sauvegarder():
    path = rag.save()
    print(f"ğŸ’¾ SauvegardÃ©: {path}")
    return path

# EXÃ‰CUTION
print("\n"+"="*80)
print("ğŸš€ DÃ‰MARRAGE")
print("="*80+"\n")

if CONFIG["groq_api_key"] == "VOTRE_CLE_API_ICI":
    print("âŒ ERREUR: Configurez clÃ© API ligne 82")
    print("   â†’ https://console.groq.com/keys")
else:
    rag = init_system()

    # TESTS
    logger.log("ğŸ§ª TESTS")

    print("\n"+"="*80)
    print("TEST: Conversation multi-tours")
    print("="*80)

    poser_question("Quels sont les Services du SecrÃ©tariat GÃ©nÃ©ral ?")
    poser_question("Comment est organisÃ© le premier ?")
    poser_question("Qui dirige l'Inspection interne ?")
    poser_question("Quelles sont ses missions ?")

    print("\n\n" + "="*80)
    print("ğŸ‰ SYSTÃˆME PRÃŠT!")
    print("="*80)
    print("""
UTILISATION:
    poser_question("Votre question")
    nouvelle_conversation()
    sauvegarder()

CAPACITÃ‰S:
    âœ¨ MÃ©moire (10 tours)
    âœ¨ CorÃ©fÃ©rences ("il", "le premier")
    âœ¨ EntitÃ©s (articles)
    âœ¨ Web fallback
""")
    print("="*80)
```

### ğŸ¤– Version 4.2 : Architecture Multi-Agents

Pour rendre le systÃ¨me robuste et modulaire, nous avons adoptÃ© une architecture multi-agents oÃ¹ chaque tÃ¢che est gÃ©rÃ©e par une entitÃ© spÃ©cialisÃ©e.

#### Les 5 Agents du SystÃ¨me :
1.  **Orchestrator Agent** : ReÃ§oit la question et coordonne les autres.
2.  **Entity Tracker Agent** : RepÃ¨re les numÃ©ros d'articles et noms de services pour rÃ©soudre les "il/elle/le".
3.  **Search Agent** : Interroge la base vectorielle ChromaDB.
4.  **Web Fallback Agent** : Va sur Google/DuckDuckGo si la base interne ne suffit pas.
5.  **Synthesis Agent** : RÃ©dige la rÃ©ponse finale propre.

Voici le code de l'agent **Entity Tracker** qui permet cette intelligence :

```python
# Code Architecture Multi-Agents
def orchestrator_agent(question, memory, entity_tracker):
    """Agent Orchestrateur - Enrichissement et rÃ©solution corÃ©fÃ©rences"""

    context = memory.get_context(n=2)

    # DÃ©tecter pronoms
    pronouns_pattern = r'\b(il|elle|le|la|ses|premier|premiÃ¨re|dernier|derniÃ¨re)\b'
    pronouns = re.findall(pronouns_pattern, question, re.IGNORECASE)

    enriched = question
    resolutions = {}

    for pronoun in pronouns:
        p = pronoun.lower()
        if p in ["premier", "premiÃ¨re"]:
            resolution = entity_tracker.get_first()
        else:
            resolution = entity_tracker.resolve(pronoun)

        if resolution and resolution != pronoun:
            enriched = enriched.replace(pronoun, resolution)
            resolutions[pronoun] = resolution

    entity_tracker.extract_entities(question + " " + enriched)

    return enriched, context, resolutions

def search_agent(query, year=None, source=None, top_k=10):
    """Agent de Recherche - Recherche sÃ©mantique dans base"""

    # Note: Utilise les variables globales 'embedding_model' et 'collection' dÃ©finies plus haut
    query_embedding = embedding_model.encode([query], convert_to_numpy=True)

    where_filter = {}
    if year and year != "Toutes":
        where_filter["year"] = str(year)
    if source:
        where_filter["source_file"] = {"$contains": source}

    results = collection_v1.query( # Utilise la collection V3 ou V4
        query_embeddings=query_embedding.tolist(),
        n_results=top_k,
        include=["documents", "metadatas", "distances"],
        where=where_filter if where_filter else None
    )

    if not results["documents"] or not results["documents"][0]:
        return [], 0.0

    docs = []
    distances = results["distances"][0]
    scores = [1 / (1 + d) for d in distances]
    confidence = sum(scores) / len(scores)

    for doc, meta, score in zip(results["documents"][0], results["metadatas"][0], scores):
        docs.append({
            "text": doc,
            "source": meta.get("source_file", "Unknown"),
            "year": meta.get("year", ""),
            "score": score,
            "metadata": meta
        })

    return docs, confidence

def web_fallback_agent(question):
    """Agent Web Fallback - Recherche DuckDuckGo si confiance faible"""
    try:
        import requests
        from bs4 import BeautifulSoup

        url = f"https://html.duckduckgo.com/html/?q={question}"
        headers = {"User-Agent": "Mozilla/5.0"}
        response = requests.get(url, headers=headers, timeout=5)

        if response.status_code == 200:
            soup = BeautifulSoup(response.text, 'html.parser')
            results = soup.find_all('a', class_='result__a', limit=3)

            web_results = []
            for result in results:
                title = result.get_text()
                link = result.get('href', '')
                web_results.append({"title": title, "link": link})
            return web_results
    except Exception as e:
        pass
    return []

def synthesis_agent(question, docs, context):
    """Agent de SynthÃ¨se - GÃ©nÃ©ration rÃ©ponse avec Groq"""
    # Note: Utilise 'groq_client' (Ã  initialiser via la classe RAG V4 plus haut ou ici)
    # Pour la dÃ©mo, on suppose 'rag.generator.client' disponible ou on recrÃ©e
    client = Groq(api_key=os.environ.get("GROQ_API_KEY", ""))

    context_str = "\n\n".join([f"[Doc {i}] {d['text'][:500]}" for i, d in enumerate(docs[:5], 1)])

    prompt = f"""Tu es un assistant juridique expert.
    SOURCES: {context_str}
    CONTEXTE: {context[:300] if context else 'Aucun'}
    QUESTION: {question}
    RÃ©ponds prÃ©cisÃ©ment."""

    try:
        completion = client.chat.completions.create(
            model="llama-3.3-70b-versatile",
            messages=[{"role": "user", "content": prompt}],
            temperature=0.3
        )
        return completion.choices[0].message.content
    except Exception as e:
        return f"âŒ Erreur Groq: {e}"

print("âœ… 4 Agents SpÃ©cialisÃ©s (Orchestrator, Search, Web, SynthÃ¨se) dÃ©finis.")
```

```python
# ğŸ’¾ RAG Indexation IncrÃ©mentale par Batches - Version 4.2
# Traite 10 PDFs Ã  la fois, sauvegarde, et reprend automatiquement

"""
â•”â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•—
â•‘    RAG INDEXATION INCRÃ‰MENTALE PAR BATCHES - v4.2                         â•‘
â• â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•£
â•‘                                                                           â•‘
â•‘  âœ¨ NOUVEAUTÃ‰S v4.2:                                                      â•‘
â•‘  âœ… Traitement par batches de 10 PDFs                                    â•‘
â•‘  âœ… Sauvegarde automatique aprÃ¨s chaque batch                            â•‘
â•‘  âœ… Reprise automatique lÃ  oÃ¹ on s'est arrÃªtÃ©                            â•‘
â•‘  âœ… RÃ©sistant aux coupures de session                                    â•‘
â•‘  âœ… Checkpoints dans Google Drive                                        â•‘
â•‘                                                                           â•‘
â•‘  ğŸ¯ USAGE:                                                                â•‘
â•‘  1. ExÃ©cuter â†’ Traite 10 PDFs (~25 min)                                 â•‘
â•‘  2. Sauvegarde automatique dans Drive                                    â•‘
â•‘  3. Session fermÃ©e ? Pas grave !                                         â•‘
â•‘  4. RÃ©-exÃ©cuter â†’ Reprend automatiquement                               â•‘
â•‘  5. RÃ©pÃ©ter jusqu'Ã  236/236 PDFs indexÃ©s                                â•‘
â•‘                                                                           â•‘
â•‘  â±ï¸ TEMPS PAR SESSION:                                                    â•‘
â•‘  â€¢ 10 PDFs : ~25 minutes (safe pour Colab gratuit)                      â•‘
â•‘  â€¢ Total : 24 sessions Ã— 25 min = Ã‰talÃ© sur plusieurs jours             â•‘
â•‘                                                                           â•‘
â•šâ•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
"""

print("ğŸ”§ Installation dÃ©pendances...")
print("="*80)

import subprocess
import sys

packages = [
    "transformers",
    "chromadb",
    "sentence-transformers",
    "chonkie",
    "unstructured[pdf]",
    "groq",
    "beautifulsoup4",
    "requests"
]

for pkg in packages:
    print(f"  Installing {pkg}...")
    try:
        subprocess.check_call([sys.executable, "-m", "pip", "install", "-q", pkg])
    except:
        print(f"  âš ï¸ Warning: {pkg} installation failed")

print("âœ… DÃ©pendances installÃ©es!\n")

# Imports
import os, re, json, hashlib, torch, requests, glob, time
from datetime import datetime
from typing import List, Dict, Optional, Tuple
from tqdm import tqdm
from collections import deque
from pathlib import Path

from unstructured.partition import pdf
from chonkie import Pipeline
from sentence_transformers import SentenceTransformer
import chromadb
from groq import Groq
from bs4 import BeautifulSoup

# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# CONFIGURATION
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

# ğŸ“ CHEMINS
FOLDER_PATH = "/content/drive/MyDrive/Datasets_journal-officiel"
CACHE_PATH = "/content/drive/MyDrive/RAG_Cache_Incremental"  # ğŸ†• Nouveau cache

CONFIG = {
    "folder_path": FOLDER_PATH,
    "cache_path": CACHE_PATH,
    "output_dir": "/content/rag_output",
    "collection_name": "journal_officiel_incremental",
    "embedding_model": "sentence-transformers/all-MiniLM-L6-v2",
    "chunk_size": 512,
    "chunk_overlap": 128,
    "pdfs_per_batch": 10,  # ğŸ†• 10 PDFs Ã  la fois !
    "max_file_size_mb": 20,
}

os.makedirs(CONFIG["output_dir"], exist_ok=True)
os.makedirs(CONFIG["cache_path"], exist_ok=True)

# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# LOGGER
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

class Logger:
    def __init__(self):
        self.log_file = os.path.join(CONFIG["output_dir"], "indexation_log.txt")

    def log(self, msg, level="INFO"):
        ts = datetime.now().strftime("%H:%M:%S")
        log = f"[{ts}] [{level}] {msg}"
        print(log)
        with open(self.log_file, "a", encoding="utf-8") as f:
            f.write(log + "\n")

logger = Logger()

# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# ğŸ†• GESTIONNAIRE DE CHECKPOINTS
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

class CheckpointManager:
    """GÃ¨re les checkpoints pour reprise automatique"""

    def __init__(self, cache_path: str):
        self.cache_path = cache_path
        self.checkpoint_file = os.path.join(cache_path, "checkpoint.json")
        self.chroma_path = os.path.join(cache_path, "chroma_db")

        os.makedirs(self.chroma_path, exist_ok=True)

    def charger_checkpoint(self) -> Dict:
        """Charger le dernier checkpoint"""
        if not os.path.exists(self.checkpoint_file):
            return {
                "pdfs_indexes": [],
                "total_chunks": 0,
                "last_batch": 0,
                "completed": False,
                "last_updated": None
            }

        try:
            with open(self.checkpoint_file, 'r', encoding='utf-8') as f:
                data = json.load(f)
            logger.log(f"âœ… Checkpoint chargÃ©: {len(data['pdfs_indexes'])} PDFs dÃ©jÃ  indexÃ©s", "SUCCESS")
            return data
        except Exception as e:
            logger.log(f"âš ï¸ Erreur lecture checkpoint: {e}", "WARNING")
            return {
                "pdfs_indexes": [],
                "total_chunks": 0,
                "last_batch": 0,
                "completed": False,
                "last_updated": None
            }

    def sauvegarder_checkpoint(self, pdfs_indexes: List[str], total_chunks: int, batch_num: int, completed: bool = False):
        """Sauvegarder checkpoint"""
        checkpoint = {
            "pdfs_indexes": pdfs_indexes,
            "total_chunks": total_chunks,
            "last_batch": batch_num,
            "completed": completed,
            "last_updated": datetime.now().isoformat()
        }

        with open(self.checkpoint_file, 'w', encoding='utf-8') as f:
            json.dump(checkpoint, f, ensure_ascii=False, indent=2)

        logger.log(f"ğŸ’¾ Checkpoint sauvegardÃ©: {len(pdfs_indexes)} PDFs, Batch {batch_num}", "SUCCESS")

    def get_pdfs_restants(self, tous_les_pdfs: List[str]) -> List[str]:
        """Obtenir liste des PDFs pas encore indexÃ©s"""
        checkpoint = self.charger_checkpoint()
        pdfs_deja_faits = set(checkpoint["pdfs_indexes"])

        # Filtrer pour obtenir seulement les non-indexÃ©s
        pdfs_restants = [p for p in tous_les_pdfs if os.path.basename(p) not in pdfs_deja_faits]

        logger.log(f"ğŸ“Š {len(pdfs_deja_faits)} PDFs dÃ©jÃ  indexÃ©s", "INFO")
        logger.log(f"ğŸ“Š {len(pdfs_restants)} PDFs restants", "INFO")

        return pdfs_restants

checkpoint_manager = CheckpointManager(CONFIG["cache_path"])

# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# EXTRACTION ET CHUNKING (identique)
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

class DocumentExtractor:
    def __init__(self, pdf_path):
        self.pdf_path = pdf_path

    def extract(self):
        try:
            document = pdf.partition_pdf(filename=self.pdf_path)
            text = " ".join([elem.text for elem in document if hasattr(elem, 'text')])
            return text
        except Exception as e:
            logger.log(f"âŒ Erreur extraction {os.path.basename(self.pdf_path)}: {e}", "ERROR")
            return ""

class IntelligentChunker:
    def __init__(self, chunk_size=512, overlap=128):
        self.chunk_size = chunk_size
        self.overlap = overlap
        self.pipeline = None

    def create_pipeline(self):
        self.pipeline = (
            Pipeline()
            .chunk_with("recursive", tokenizer="roberta-base", chunk_size=2048, recipe="markdown")
            .chunk_with("semantic", chunk_size=self.chunk_size)
            .refine_with("overlap", context_size=self.overlap)
            .refine_with("embeddings", embedding_model=CONFIG["embedding_model"])
        )

    def chunk_text(self, text):
        if not self.pipeline:
            self.create_pipeline()

        if not text or len(text) < 50:
            return []

        try:
            doc = self.pipeline.run(texts=text)
            chunks = [chunk.text for chunk in doc.chunks]
            return chunks
        except Exception as e:
            logger.log(f"âš ï¸ Chunking error: {e}", "WARNING")
            return []

class MetadataExtractor:
    @staticmethod
    def extract_date_from_filename(filename: str) -> Optional[str]:
        patterns = [
            r'(\d{2})\.(\d{2})\.(\d{4})',
            r'DU[_\s-]+(\d{2})\.(\d{2})\.(\d{4})',
        ]

        for pattern in patterns:
            match = re.search(pattern, filename)
            if match:
                if len(match.groups()) == 3:
                    day, month, year = match.groups()
                    try:
                        return f"{year}-{month}-{day}"
                    except:
                        pass
        return None

    @staticmethod
    def extract(chunk_text: str, chunk_index: int, source_doc: str) -> Dict:
        metadata = {
            "chunk_index": chunk_index,
            "source_file": source_doc,
        }

        doc_date = MetadataExtractor.extract_date_from_filename(source_doc)
        if doc_date:
            metadata["document_date"] = doc_date
            metadata["year"] = doc_date.split('-')[0]
        else:
            metadata["year"] = ""

        return metadata

# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# ğŸ†• VECTOR STORE INCRÃ‰MENTAL
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

class IncrementalVectorStore:
    """Vector store avec sauvegarde incrÃ©mentale"""

    def __init__(self, collection_name, embedding_model_name, chroma_path):
        self.collection_name = collection_name
        self.embedding_model_name = embedding_model_name
        self.chroma_path = chroma_path
        self.client = None
        self.collection = None
        self.embedding_model = None

    def initialize(self):
        """Initialiser ou charger collection"""
        logger.log("ğŸ”Œ Initialisation Vector Store...", "INFO")

        # Client avec persistance
        self.client = chromadb.PersistentClient(path=self.chroma_path)

        try:
            # Essayer de charger existante
            self.collection = self.client.get_collection(name=self.collection_name)
            count = self.collection.count()
            logger.log(f"âœ… Collection chargÃ©e: {count} documents", "SUCCESS")
        except:
            # CrÃ©er nouvelle
            self.collection = self.client.create_collection(name=self.collection_name)
            logger.log("âœ… Nouvelle collection crÃ©Ã©e", "SUCCESS")

        # Charger modÃ¨le embedding
        self.embedding_model = SentenceTransformer(self.embedding_model_name)
        logger.log("âœ… ModÃ¨le embedding chargÃ©", "SUCCESS")

    def add_documents(self, chunks: List[str], source_document: str):
        """Ajouter documents et sauvegarder automatiquement"""
        if not chunks:
            return 0

        logger.log(f"ğŸ“Š Ajout {len(chunks)} chunks de {source_document[:50]}...", "INFO")

        # Calculer embeddings
        embeddings = self.embedding_model.encode(chunks, show_progress_bar=False, convert_to_numpy=True)

        doc_hash = hashlib.md5(source_document.encode()).hexdigest()[:8]

        # Ajouter Ã  ChromaDB
        for i, (chunk, embedding) in enumerate(zip(chunks, embeddings)):
            metadata = MetadataExtractor.extract(chunk, i, source_document)
            unique_id = f"{doc_hash}_{i}"

            try:
                self.collection.add(
                    documents=[chunk],
                    embeddings=[embedding.tolist()],
                    metadatas=[metadata],
                    ids=[unique_id]
                )
            except Exception as e:
                logger.log(f"âš ï¸ Erreur ajout chunk {i}: {e}", "WARNING")

        # ChromaDB sauvegarde automatiquement (persistance)
        return len(chunks)

# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# ğŸ†• INDEXATION PAR BATCHES
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

def scanner_pdfs(folder_path: str, max_size_mb: int = 20) -> List[str]:
    """Scanner tous les PDFs valides"""
    logger.log(f"ğŸ” Scan dossier: {folder_path}", "INFO")

    pdf_pattern = os.path.join(folder_path, "*.pdf")
    all_pdfs = glob.glob(pdf_pattern)

    valid_pdfs = []
    skipped = 0

    for pdf in all_pdfs:
        try:
            size_mb = os.path.getsize(pdf) / (1024 * 1024)
            if size_mb <= max_size_mb:
                valid_pdfs.append(pdf)
            else:
                logger.log(f"âš ï¸ Trop gros: {os.path.basename(pdf)} ({size_mb:.1f} MB)", "WARNING")
                skipped += 1
        except Exception as e:
            logger.log(f"âš ï¸ Erreur: {os.path.basename(pdf)}", "WARNING")
            skipped += 1

    logger.log(f"âœ… {len(valid_pdfs)} PDFs valides ({skipped} ignorÃ©s)", "SUCCESS")
    return sorted(valid_pdfs)

def indexer_batch(pdfs_batch: List[str], batch_num: int, chunker: IntelligentChunker, vector_store: IncrementalVectorStore) -> Tuple[List[str], int]:
    """
    Indexer un batch de PDFs

    Returns:
        (pdfs_indexes_avec_succÃ¨s, total_chunks)
    """
    logger.log(f"\n{'='*80}", "INFO")
    logger.log(f"ğŸ“¦ BATCH {batch_num} - {len(pdfs_batch)} PDFs", "INFO")
    logger.log(f"{'='*80}", "INFO")

    pdfs_indexes = []
    total_chunks = 0

    for i, pdf_path in enumerate(pdfs_batch, 1):
        filename = os.path.basename(pdf_path)
        logger.log(f"\nğŸ“„ [{i}/{len(pdfs_batch)}] {filename}", "INFO")

        try:
            # Extraction
            extractor = DocumentExtractor(pdf_path)
            text = extractor.extract()

            if not text:
                logger.log(f"  âš ï¸ Texte vide", "WARNING")
                continue

            logger.log(f"  âœ… Extrait: {len(text)} caractÃ¨res", "SUCCESS")

            # Chunking
            chunks = chunker.chunk_text(text)

            if not chunks:
                logger.log(f"  âš ï¸ Aucun chunk", "WARNING")
                continue

            logger.log(f"  âœ… ChunkÃ©: {len(chunks)} chunks", "SUCCESS")

            # Indexation
            chunks_added = vector_store.add_documents(chunks, filename)

            logger.log(f"  âœ… IndexÃ©: {filename}", "SUCCESS")

            # SuccÃ¨s !
            pdfs_indexes.append(filename)
            total_chunks += chunks_added

        except Exception as e:
            logger.log(f"  âŒ Erreur: {e}", "ERROR")
            continue

    logger.log(f"\n{'='*80}", "INFO")
    logger.log(f"âœ… Batch {batch_num} terminÃ©: {len(pdfs_indexes)}/{len(pdfs_batch)} PDFs indexÃ©s", "SUCCESS")
    logger.log(f"ğŸ“Š {total_chunks} chunks ajoutÃ©s", "SUCCESS")
    logger.log(f"{'='*80}\n", "INFO")

    return pdfs_indexes, total_chunks

def process_incremental():
    """ğŸ†• Processus principal incrÃ©mental"""

    logger.log("\n" + "="*80, "INFO")
    logger.log("ğŸš€ INDEXATION INCRÃ‰MENTALE PAR BATCHES", "INFO")
    logger.log("="*80 + "\n", "INFO")

    # 1. Scanner tous les PDFs
    tous_les_pdfs = scanner_pdfs(CONFIG["folder_path"], CONFIG["max_file_size_mb"])

    if not tous_les_pdfs:
        logger.log("âŒ Aucun PDF trouvÃ©", "ERROR")
        return

    # 2. Charger checkpoint et obtenir PDFs restants
    pdfs_restants = checkpoint_manager.get_pdfs_restants(tous_les_pdfs)

    if not pdfs_restants:
        logger.log("ğŸ‰ TOUS LES PDFs DÃ‰JÃ€ INDEXÃ‰S !", "SUCCESS")
        checkpoint_manager.sauvegarder_checkpoint(
            [os.path.basename(p) for p in tous_les_pdfs],
            checkpoint_manager.charger_checkpoint()["total_chunks"],
            checkpoint_manager.charger_checkpoint()["last_batch"],
            completed=True
        )
        return

    # 3. Prendre seulement les N premiers (10 par dÃ©faut)
    pdfs_ce_batch = pdfs_restants[:CONFIG["pdfs_per_batch"]]

    logger.log(f"\n{'='*80}", "INFO")
    logger.log(f"ğŸ“Š PLAN D'INDEXATION", "INFO")
    logger.log(f"{'='*80}", "INFO")
    logger.log(f"Total PDFs: {len(tous_les_pdfs)}", "INFO")
    logger.log(f"DÃ©jÃ  indexÃ©s: {len(tous_les_pdfs) - len(pdfs_restants)}", "INFO")
    logger.log(f"Restants: {len(pdfs_restants)}", "INFO")
    logger.log(f"Ce batch: {len(pdfs_ce_batch)}", "INFO")
    logger.log(f"{'='*80}\n", "INFO")

    # 4. Initialiser chunker et vector store
    logger.log("âœ‚ï¸ CrÃ©ation chunker...", "INFO")
    chunker = IntelligentChunker(CONFIG["chunk_size"], CONFIG["chunk_overlap"])
    logger.log("âœ… Chunker OK", "SUCCESS")

    vector_store = IncrementalVectorStore(
        CONFIG["collection_name"],
        CONFIG["embedding_model"],
        checkpoint_manager.chroma_path
    )
    vector_store.initialize()

    # 5. Indexer ce batch
    checkpoint = checkpoint_manager.charger_checkpoint()
    batch_num = checkpoint["last_batch"] + 1

    pdfs_indexes, chunks_ajoutes = indexer_batch(
        pdfs_ce_batch,
        batch_num,
        chunker,
        vector_store
    )

    # 6. Sauvegarder checkpoint
    tous_pdfs_indexes = checkpoint["pdfs_indexes"] + pdfs_indexes
    total_chunks_global = checkpoint["total_chunks"] + chunks_ajoutes

    completed = (len(tous_pdfs_indexes) >= len(tous_les_pdfs))

    checkpoint_manager.sauvegarder_checkpoint(
        tous_pdfs_indexes,
        total_chunks_global,
        batch_num,
        completed=completed
    )

    # 7. RÃ©sumÃ©
    logger.log(f"\n{'='*80}", "INFO")
    logger.log(f"ğŸ“Š RÃ‰SUMÃ‰ SESSION", "INFO")
    logger.log(f"{'='*80}", "INFO")
    logger.log(f"PDFs indexÃ©s ce batch: {len(pdfs_indexes)}/{len(pdfs_ce_batch)}", "INFO")
    logger.log(f"Chunks ajoutÃ©s: {chunks_ajoutes}", "INFO")
    logger.log(f"Progression totale: {len(tous_pdfs_indexes)}/{len(tous_les_pdfs)} PDFs", "INFO")
    logger.log(f"Chunks totaux: {total_chunks_global}", "INFO")

    if completed:
        logger.log(f"\nğŸ‰ INDEXATION COMPLÃˆTE TERMINÃ‰E !", "SUCCESS")
    else:
        restants = len(tous_les_pdfs) - len(tous_pdfs_indexes)
        logger.log(f"\nâ­ï¸ PROCHAINE SESSION: {restants} PDFs restants", "INFO")
        logger.log(f"ğŸ’¡ RÃ©-exÃ©cutez ce code pour continuer", "INFO")

    logger.log(f"{'='*80}\n", "INFO")

    return {
        "pdfs_indexes_ce_batch": len(pdfs_indexes),
        "chunks_ajoutes": chunks_ajoutes,
        "total_pdfs": len(tous_pdfs_indexes),
        "total_chunks": total_chunks_global,
        "completed": completed
    }

# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# EXÃ‰CUTION
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

print("\n" + "="*80)
print("ğŸš€ DÃ‰MARRAGE INDEXATION INCRÃ‰MENTALE v4.2")
print("="*80 + "\n")

# VÃ©rifier dossier
if not os.path.exists(CONFIG["folder_path"]):
    print(f"âŒ Dossier non trouvÃ©: {CONFIG['folder_path']}")
    print("ğŸ’¡ Montez Google Drive et vÃ©rifiez le chemin")
else:
    # Lancer processus
    resultat = process_incremental()

    if resultat:
        print("\n" + "="*80)
        print("âœ… SESSION TERMINÃ‰E")
        print("="*80)
        print(f"ğŸ“Š Progression: {resultat['total_pdfs']} PDFs indexÃ©s")
        print(f"ğŸ“Š Total chunks: {resultat['total_chunks']}")

        if resultat['completed']:
            print("\nğŸ‰ INDEXATION COMPLÃˆTE ! SystÃ¨me prÃªt Ã  l'emploi")
        else:
            print(f"\nâ­ï¸ RÃ©-exÃ©cutez pour continuer l'indexation")

        print("="*80)
```

```python
# ğŸ” RAG Query - Interroger la Base IndexÃ©e

"""
â•”â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•—
â•‘    MODULE D'INTERROGATION - RAG v4.2                                      â•‘
â• â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•£
â•‘                                                                           â•‘
â•‘  âœ¨ PRÃ‰REQUIS:                                                            â•‘
â•‘  âœ… Indexation terminÃ©e (224 PDFs, 41,922 chunks)                        â•‘
â•‘  âœ… Cache dans /Drive/RAG_Cache_Incremental/                             â•‘
â•‘                                                                           â•‘
â•‘  ğŸ¯ FONCTIONNALITÃ‰S:                                                      â•‘
â•‘  âœ… Recherche sÃ©mantique dans 224 PDFs                                   â•‘
â•‘  âœ… Filtres annÃ©e et source                                              â•‘
â•‘  âœ… MÃ©moire conversationnelle                                            â•‘
â•‘  âœ… RÃ©solution corÃ©fÃ©rences                                              â•‘
â•‘  âœ… Groq LLM pour synthÃ¨se                                               â•‘
â•‘                                                                           â•‘
â•‘  âš¡ USAGE:                                                                 â•‘
â•‘  1. Configurer clÃ© API Groq (ligne 30)                                  â•‘
â•‘  2. ExÃ©cuter ce code                                                     â•‘
â•‘  3. Utiliser poser_question("...")                                      â•‘
â•‘                                                                           â•‘
â•šâ•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
"""

import os, json, time, re
from datetime import datetime
from typing import List, Dict, Optional
from collections import deque

import chromadb
from sentence_transformers import SentenceTransformer
from groq import Groq

# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# CONFIGURATION
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

# ğŸ”‘ CLÃ‰ API GROQ (OBLIGATOIRE pour rÃ©ponses)
#GROQ_API_KEY = "VOTRE_CLE_API_ICI"  # â† CHANGEZ ICI !

# ğŸ“ CHEMIN CACHE (mÃªme que indexation)
CACHE_PATH = "/content/drive/MyDrive/RAG_Cache_Incremental"

CONFIG = {
    "cache_path": CACHE_PATH,
    "collection_name": "journal_officiel_incremental",
    "embedding_model": "sentence-transformers/all-MiniLM-L6-v2",
    "groq_api_key": GROQ_API_KEY,
    "groq_model": "llama-3.3-70b-versatile",
    "top_k": 10,
    "memory_turns": 10,
}

# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# MÃ‰MOIRE CONVERSATIONNELLE
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

class ConversationMemory:
    def __init__(self, max_turns=10):
        self.max_turns = max_turns
        self.history = deque(maxlen=max_turns)

    def add(self, question, answer):
        self.history.append({
            "question": question,
            "answer": answer,
            "timestamp": datetime.now().isoformat()
        })

    def get_context(self, n=3):
        recent = list(self.history)[-n:]
        context = []
        for turn in recent:
            context.append(f"Q: {turn['question']}\nR: {turn['answer'][:200]}")
        return "\n\n".join(context)

    def clear(self):
        self.history.clear()

class EntityTracker:
    def __init__(self):
        self.last_mentioned = {}
        self.mention_order = []

    def extract_entities(self, text):
        # Articles
        articles = re.findall(r'[Aa]rticle\s+(\d+)', text)
        if articles:
            self.last_mentioned["article"] = articles[-1]
            self.mention_order.append(("article", articles[-1]))

        # Directions/Services
        directions = re.findall(r'(?:Direction|Service)\s+(?:de\s+)?(?:la\s+)?([A-ZÃ€-Ã¿\s]+)', text)
        cleaned = [d.strip() for d in directions if len(d.strip()) > 3]
        if cleaned:
            self.last_mentioned["direction"] = cleaned[-1]
            self.mention_order.append(("direction", cleaned[-1]))

    def resolve(self, pronoun):
        p = pronoun.lower()
        if p in ["il", "elle", "le", "la"]:
            for etype in ["article", "direction"]:
                if etype in self.last_mentioned:
                    return f"{etype.capitalize()} {self.last_mentioned[etype]}"
        return pronoun

    def get_first(self):
        if self.mention_order:
            etype, value = self.mention_order[0]
            return f"{etype.capitalize()} {value}"
        return ""

# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# VECTOR STORE QUERY
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

class VectorStoreQuery:
    """Interroger la base vectorielle indexÃ©e"""

    def __init__(self, cache_path, collection_name, embedding_model_name):
        self.cache_path = cache_path
        self.collection_name = collection_name
        self.embedding_model_name = embedding_model_name

        print("ğŸ”Œ Chargement base vectorielle...")

        # Client ChromaDB
        chroma_path = os.path.join(cache_path, "chroma_db")
        self.client = chromadb.PersistentClient(path=chroma_path)

        # Collection
        try:
            self.collection = self.client.get_collection(name=collection_name)
            count = self.collection.count()
            print(f"âœ… Collection chargÃ©e: {count:,} documents")
        except Exception as e:
            print(f"âŒ Erreur: Collection non trouvÃ©e")
            print(f"ğŸ’¡ Assurez-vous d'avoir exÃ©cutÃ© RAG_INCREMENTAL.py d'abord")
            raise e

        # ModÃ¨le embedding
        print("ğŸ”Œ Chargement modÃ¨le embedding...")
        self.embedding_model = SentenceTransformer(embedding_model_name)
        print("âœ… ModÃ¨le chargÃ©")

    def search(self, query: str, n_results: int = 10, year: Optional[str] = None, source: Optional[str] = None):
        """Recherche sÃ©mantique"""

        # Embedding de la question
        query_embedding = self.embedding_model.encode([query], convert_to_numpy=True)

        # Filtres
        where_filter = {}
        if year:
            where_filter["year"] = year
        if source:
            where_filter["source_file"] = {"$contains": source}

        # Recherche
        try:
            results = self.collection.query(
                query_embeddings=query_embedding.tolist(),
                n_results=n_results,
                include=['documents', 'metadatas', 'distances'],
                where=where_filter if where_filter else None
            )
        except Exception as e:
            print(f"âŒ Erreur recherche: {e}")
            return [], 0.0

        if not results['documents'][0]:
            return [], 0.0

        # Calculer scores
        distances = results['distances'][0]
        scores = [1 / (1 + d) for d in distances]
        confidence = sum(scores) / len(scores)

        # Formater rÃ©sultats
        formatted = []
        for doc, metadata, score in zip(results['documents'][0], results['metadatas'][0], scores):
            formatted.append({
                'text': doc,
                'metadata': metadata,
                'score': score,
                'source': metadata.get('source_file', 'Unknown')
            })

        return formatted, confidence

# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# AGENTS RAG
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

def orchestrator_agent(question, memory, entity_tracker):
    """Enrichir question avec contexte et rÃ©soudre corÃ©fÃ©rences"""

    context = memory.get_context(n=2)

    # DÃ©tecter pronoms et rÃ©fÃ©rences
    pronouns_pattern = r'\b(il|elle|le|la|ses|premier|premiÃ¨re|dernier|derniÃ¨re)\b'
    pronouns = re.findall(pronouns_pattern, question, re.IGNORECASE)

    enriched = question
    resolutions = {}

    for pronoun in pronouns:
        p = pronoun.lower()
        if p in ["premier", "premiÃ¨re"]:
            resolution = entity_tracker.get_first()
        else:
            resolution = entity_tracker.resolve(pronoun)

        if resolution and resolution != pronoun:
            enriched = enriched.replace(pronoun, resolution)
            resolutions[pronoun] = resolution

    entity_tracker.extract_entities(question + " " + enriched)

    return enriched, context, resolutions

def synthesis_agent(question, sources, context, groq_client):
    """GÃ©nÃ©rer rÃ©ponse avec Groq"""

    # PrÃ©parer contexte
    contexts = []
    for i, src in enumerate(sources[:5], 1):
        source_info = f"Source: {src.get('source', 'Unknown')}"
        contexts.append(f"[Doc {i}] {source_info}\n{src['text'][:500]}")

    context_str = "\n\n".join(contexts)

    # Prompt
    prompt = f"""Tu es un assistant juridique expert.

SOURCES DOCUMENTAIRES:
{context_str}

CONTEXTE CONVERSATIONNEL:
{context[:300] if context else 'Aucun'}

QUESTION: {question}

INSTRUCTIONS:
- RÃ©ponds de maniÃ¨re prÃ©cise et structurÃ©e
- Utilise des points â€¢ pour les listes
- Cite les sources (fichiers) quand pertinent
- Si info manquante, indique-le clairement

RÃ‰PONSE:"""

    try:
        completion = groq_client.chat.completions.create(
            model=CONFIG["groq_model"],
            messages=[{"role": "user", "content": prompt}],
            temperature=0.3
        )
        answer = completion.choices[0].message.content
        return answer
    except Exception as e:
        return f"âŒ Erreur gÃ©nÃ©ration rÃ©ponse: {e}\nğŸ’¡ VÃ©rifiez votre clÃ© API Groq"

# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# FONCTION PRINCIPALE
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

# Initialisation
print("\n" + "="*80)
print("ğŸš€ INITIALISATION SYSTÃˆME D'INTERROGATION")
print("="*80 + "\n")

# Vector Store
vector_store = VectorStoreQuery(
    CONFIG["cache_path"],
    CONFIG["collection_name"],
    CONFIG["embedding_model"]
)

# MÃ©moire
conversation_memory = ConversationMemory(max_turns=CONFIG["memory_turns"])
entity_tracker = EntityTracker()

# Client Groq
if CONFIG["groq_api_key"] and CONFIG["groq_api_key"] != "VOTRE_CLE_API_ICI":
    groq_client = Groq(api_key=CONFIG["groq_api_key"])
    print("âœ… Client Groq initialisÃ©")
else:
    groq_client = None
    print("âš ï¸ ClÃ© API Groq non configurÃ©e")
    print("ğŸ’¡ Configurez ligne 30 pour obtenir des rÃ©ponses")

print("\n" + "="*80)
print("âœ… SYSTÃˆME PRÃŠT !")
print("="*80)
print(f"ğŸ“Š Base: 224 PDFs, 41,922 chunks")
print(f"ğŸ” Utilisez: poser_question('votre question')")
print("="*80 + "\n")

# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# FONCTION UTILISATEUR
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

def poser_question(question: str, year: Optional[str] = None, source: Optional[str] = None):
    """
    Poser une question au systÃ¨me RAG

    Args:
        question: Question Ã  poser
        year: Filtre annÃ©e optionnel (ex: "2023")
        source: Filtre source optionnel (ex: "1389")

    Returns:
        dict avec answer, confidence, sources
    """

    print("\n" + "="*80)
    print(f"â“ {question}")
    if year:
        print(f"ğŸ“… Filtre annÃ©e: {year}")
    if source:
        print(f"ğŸ“„ Filtre source: {source}")
    print("="*80)

    start_time = time.time()

    # 1. Orchestration (enrichissement + corÃ©fÃ©rences)
    enriched_q, context, resolutions = orchestrator_agent(question, conversation_memory, entity_tracker)

    if resolutions:
        print(f"ğŸ”„ CorÃ©fÃ©rences rÃ©solues: {resolutions}")

    # 2. Recherche
    print("ğŸ” Recherche dans la base...")
    results, confidence = vector_store.search(enriched_q, n_results=CONFIG["top_k"], year=year, source=source)

    print(f"ğŸ“Š Confiance: {confidence:.2%} ({len(results)} rÃ©sultats)")

    if not results:
        print("âŒ Aucun rÃ©sultat trouvÃ©")
        return {
            "answer": "DÃ©solÃ©, je n'ai pas trouvÃ© d'information pertinente dans les documents.",
            "confidence": 0.0,
            "sources": []
        }

    # 3. SynthÃ¨se
    if groq_client:
        print("âœï¸ GÃ©nÃ©ration rÃ©ponse...")
        answer = synthesis_agent(question, results, context, groq_client)
    else:
        # Mode sans Groq: retourner chunks bruts
        answer = "âš ï¸ Mode sans Groq (configurez clÃ© API pour rÃ©ponses synthÃ©tisÃ©es)\n\n"
        answer += "ğŸ“„ Extraits pertinents:\n\n"
        for i, r in enumerate(results[:3], 1):
            answer += f"{i}. {r['text'][:300]}...\n\n"

    # 4. Sauvegarder dans mÃ©moire
    conversation_memory.add(question, answer)
    entity_tracker.extract_entities(question + " " + answer)

    # 5. Affichage
    exec_time = time.time() - start_time

    print(f"\nğŸ’¬ RÃ‰PONSE:\n{answer}\n")
    print(f"ğŸ“Š Confiance: {confidence:.2%}")
    print(f"ğŸ“š Sources: {len(results)}")

    # Sources uniques
    sources_uniques = []
    seen = set()
    for r in results:
        src = r.get('source', 'Unknown')
        if src not in seen:
            seen.add(src)
            sources_uniques.append(src)

    if sources_uniques:
        print(f"\nğŸ“„ DOCUMENTS SOURCES:")
        for i, src in enumerate(sources_uniques[:5], 1):
            print(f"  {i}. {src}")

    print(f"\nâ±ï¸  Temps: {exec_time:.2f}s")
    print("="*80)

    return {
        "answer": answer,
        "confidence": confidence,
        "sources": results
    }

def nouvelle_conversation():
    """RÃ©initialiser la mÃ©moire"""
    global entity_tracker
    conversation_memory.clear()
    entity_tracker = EntityTracker()
    print("ğŸ”„ Nouvelle conversation - MÃ©moire rÃ©initialisÃ©e")

# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# TESTS (DÃ©commenter pour tester)
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

# Test simple
# poser_question("Quels sont les Services du SecrÃ©tariat GÃ©nÃ©ral ?")

# Test avec filtre
# poser_question("Quelles lois ont Ã©tÃ© votÃ©es ?", year="2023")

# Test corÃ©fÃ©rence
# poser_question("Quels sont les Services ?")
# poser_question("DÃ©taille le premier")

print("""
â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
ğŸ¯ EXEMPLES D'UTILISATION

# Question simple
poser_question("Quels sont les Services du SecrÃ©tariat GÃ©nÃ©ral ?")

# Avec filtre annÃ©e
poser_question("Quelles rÃ©formes en 2023 ?", year="2023")

# Avec filtre source
poser_question("Documents sur Budget", source="Budget")

# Conversation avec mÃ©moire
poser_question("Quels sont les Services ?")
poser_question("DÃ©taille le premier")  # CorÃ©fÃ©rence

# Nouvelle conversation
nouvelle_conversation()

â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
""")
```
